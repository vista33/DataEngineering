# Part A
### Basic mathematic operations

```python
# Make two variables
a = 12
b = 5

# 1
# addition between two variables
total = a + b
print (f'a + b = {total}')

# subraction between two variables
subtracting = a - b
print (f'a - b = {subtracting}')

# 2
# multiplication between two variables
multiplication = a * b
print (f'a x b = {multiplication}')

# division between two variables
division = a / b
print (f'a : b = {division}')

# 3
# floor division between two variables
floor_division = a // b
print (f'a // b = {floor_division}')

# modulus between two variables
modulus = a % b
print (f'a % b = {modulus}')

# 4
# exponent between two variables
exponent = a ** b
print (f'a ^ b = {exponent}')
```

# Part C
### Data Frame Exploration and Manipulation


```python
# 1. Identify continue and categorical data
# Import pandas library
import pandas as pd

# Read csv as a dataframe
df = pd.read_csv ('/content/train_and_test.csv')
print (df.columns)

# Clean the data
df = df [['Passengerid', 'Age', 'Fare', 'Sex', 'sibsp', 'Pclass', 'Embarked', '2urvived']]
df = df.rename (columns = {'2urvived':'Survived'})
df.head(5)
```

### Information about the parameters in dataframe
1. Passengerid = Unique identifier for each passenger
2. Age = Age of passenger
3. Fare = Ticket fare paid
4. Sex = Passenger's gender
5. sibsp = Number of siblings/spouses aboard the Titanic with the passenger
6. Parch = Number of parents/children aboard Titanic with the passenger
7. Pclass = Ticket class (1st, 2nd, 3rd)
8. Embarked = Port of embarkation
9. Survived = Survived pasenger (0 = did not survive, 1 = survived)


Based on the information of the parameters in the dataframe,

Based on the columns information:
1.  **Passengerid**, **Sex** and **Embarked** are qualitative data and categorized as **nominal data** because they show the label / identity and can not be compared between one and another.
2. **Pclass** and **Survived** are qualitative data and categorized as **ordinal data** because they can be sorted or ranked.
3. **Age** and **Fare** are quantitative data and categorized as **continuos data** because they can be measured and can be in decimal form.
4. **Parch** and **sibsp** are quantitative data and categorized as **discrete data** because they can be counted and have no decimal form.

```python
# 2. Looping implementation in dataframe
# Count total children, adult, and elder using for loop
total_children = 0
total_adult = 0
total_elderly = 0
for i in df['Age']:
  if i <= 17:
    total_children += 1
  elif i <= 60:
    total_adult +=1
  else:
    total_elderly += 1

print (f'Total passenger who are children {total_children}, adult {total_adult}, and elder {total_elderly}')

# Make a function to categorize age
def category_age (x):
  if x <= 17:
    y = "Children"
  elif x <= 60:
    y = "Adult"
  else:
    y = "Elderly"
  return y

# Manipulate data by making a new column to show the age category
df['age_category'] = df['Age'].map (category_age)

# Count the survived passenger using while loop
n = 0
while i < len (df):
  if df.loc[i,'Survived'] == 1:
    n += 1
  i += 1
print (f'Total survived passengers is {n}.')
```
