```python
# Learn If-Condition
blocked_rek = [123,456,789]
rek = 456

if rek not in blocked_rek:
  print('Transaction is success')
else:
  print ('Your acount is illegal')
```

```python
# Learn For-Loop
list_number = range (1,33)
for x in list_number:
  if x % 2 == 0:
    print (x, 'Even number')
  else:
    print (x, 'Odd number')
```

```python
# Latihan Santuy
# Make input from the user
print ('Please provide the temperature in Celcius:')
temp_c = float (input())

# Convert C to F
temp_f = (temp_c*9 /5) +32

# If-Logic
if temp_f > 150:
  indicator = 'Now, the temperature is hot'
elif temp_f > 100:
  indicator = 'Now, the temperature is normal'
else:
  indicator = 'Now, the temperature is cold'
print (indicator)
```
